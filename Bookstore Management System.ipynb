```python


class Book:
    def __init__(self, title, author, price):
        self.title = title
        self.author = author
        self.price = price


class Bookstore:
    def __init__(self):
        self.books = []

    def add_book(self):
        title = input("Enter book title: ")
        author = input("Enter author name: ")
        price = float(input("Enter book price: ‚Çπ"))
        new_book = Book(title, author, price)
        self.books.append(new_book)
        print(f"\n‚úÖ '{title}' added successfully!\n")

    def view_books(self):
        if not self.books:
            print("\nüì≠ No books available.\n")
        else:
            print("\nüìö Available Books:")
            for i, book in enumerate(self.books, 1):
                print(f"{i}. {book.title} by {book.author} - ‚Çπ{book.price}")
            print()

    def search_book(self):
        keyword = input("Enter book title or author to search: ").lower()
        found = [book for book in self.books if keyword in book.title.lower() or keyword in book.author.lower()]

        if found:
            print("\nüîç Search Results:")
            for book in found:
                print(f"{book.title} by {book.author} - ‚Çπ{book.price}")
        else:
            print("\n‚ùå No matching book found.\n")

    def delete_book(self):
        self.view_books()
        if self.books:
            try:
                index = int(input("Enter book number to delete: ")) - 1
                if 0 <= index < len(self.books):
                    removed = self.books.pop(index)
                    print(f"\nüóëÔ∏è '{removed.title}' deleted successfully!\n")
                else:
                    print("\n‚ö†Ô∏è Invalid book number.\n")
            except ValueError:
                print("\n‚ö†Ô∏è Please enter a valid number.\n")


def main():
    store = Bookstore()

    while True:
        print("===== üìò Bookstore Management System =====")
        print("1. Add Book")
        print("2. View All Books")
        print("3. Search Book")
        print("4. Delete Book")
        print("5. Exit")

        choice = input("Enter your choice (1-5): ")

        if choice == "1":
            store.add_book()
        elif choice == "2":
            store.view_books()
        elif choice == "3":
            store.search_book()
        elif choice == "4":
            store.delete_book()
        elif choice == "5":
            print("\nüëã Thank you for using Bookstore System! Bye!\n")
            break
        else:
            print("\n‚ö†Ô∏è Invalid choice, please try again.\n")


# Run the program
main()
```
